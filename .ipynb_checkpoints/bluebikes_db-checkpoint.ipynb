```python
import pandas as pd
import sqlite3
```

```python
db = sqlite3.connect("../bluebikes.db")
```

### Data was downloaded from csv.file off of bluebikes aka hubway website ( https://s3.amazonaws.com/hubway-data/index.html  ) and loaded into sqlite3 db

```python
def run_query(query, database):
    return pd.read_sql_query(query,database)
```

```python

query = """SELECT * FROM bluebikes LIMIT 5;"""
```

```python
run_query(query, db)
```

```python
query = """SELECT b.tripduration, b.starttime, b.start_station_name FROM bluebikes b ORDER BY b.tripduration DESC LIMIT 5;"""
```

```python
run_query(query, db)
```

```python
query = '''
SELECT COUNT(*) AS "Total Trips by Registered Users"
FROM bluebikes
WHERE usertype = "Subscriber";
'''
```

```python
run_query(query, db)
```

```python
query = """SELECT * FROM bluebikes;""" 
bikes_df = run_query(query, db)
```
